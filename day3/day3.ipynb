Today, I learned NumPy arrays and benchmark. I compared two programs that achieve the same result: one using a simple Python loop, and one using a NumPy array.

```python
import numpy as np
def py_square(n):
    return [i*i for i in range(n)]
```

```python
time(py_square(10000000))
```

```python
def numpy_square(n):
    return np.square(np.arange(n))
```

```python
time (numpy_square(10000000))
```

NumPy is as much as 30 times more efficient than a regular python loop.

I learned about parameterized functions and experimented with my sin and cos functions from yesterday to make a lissajous pattern:

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = plt.axes()

t = np.linspace(np.pi * 0.5, np.pi * 6.5, 400)

x = np.cos(t + 0.261799)
y = np.sin(t * 4)

ax.plot(x, y, 'steelblue', label='sin(x)');

plt.legend();
plt.show()
```
