Today, I learned about matplotlib and NumPy.

I wrote the sin and cos function into a graph:

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
x2 = np.linspace(0, 10, 1000)

ax.plot(x, np.cos(x), 'chocolate', label='sin(x)');
ax.plot(x2, np.sin(x2), 'teal', label='cos(x)');

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)

plt.legend();
```
